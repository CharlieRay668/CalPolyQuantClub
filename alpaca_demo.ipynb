```python
import os
import pandas as pd
import plotly.graph_objects as go
import plotly.offline as ofl
from datetime import datetime
import requests
import json
from dotenv import load_dotenv
# Load in our alpaca environment variables
load_dotenv()
```

```python
# Get the environment variables from the .env file
secret_key = os.getenv("ALPACA_SECRET_KEY")
public_key = os.getenv("ALPACA_PUBLIC_KEY")
```

```python
# Set the headers for the API request, This authenticates us, and the API keys are essentially just passwords
headers = {
    "APCA-API-KEY-ID": public_key,
    "APCA-API-SECRET-KEY": secret_key
}
# This is the API URL we are going to be using
url = "https://data.alpaca.markets/v2/stocks/bars"
# Set the parameters for the API request, defines what data we want to get
# More information can be found here, https://docs.alpaca.markets/reference/stockbars
params = {
    "limit": 1000,
    # Do not adjust for splits
    "adjustment": "raw",
    # Get Apple Data
    "symbols": "AAPL",
    # Get 1 minute bars
    "timeframe": "1T",
    # Sets our start and stop timeframes,
    # We could get more granular however I just have it grab a whole day
    "start": "2023-05-29",
    "end": "2023-05-30"
}
# Make the API request.
response = requests.get(url, headers=headers, params=params)
# This code will get very repetative, eventually the quant club will make a wrapper for this
# However for now we will just use the json library to parse the response
response
```

```python
# The data is returned to us as JSON, parse it as such
data = response.json()
# Get the bars from the JSON
bars = data["bars"]["AAPL"]
# Create the dataframe from the bars
# Dataframes are tables, you can think of them as excel sheets
df = pd.DataFrame.from_records(bars)
df.head()
```

```python
# Create a candlestick chart using plotly
fig = go.Figure(data=[go.Candlestick(x=df['t'],
                open=df['o'],
                high=df['h'],
                low=df['h'],
                close=df['c'])])
```

```python
# Show the chart
ofl.iplot(fig)
# All of our data is in UTC time, 
# For reference, market hours are 9:30 AM - 4:00 PM EST, which is 13:30 - 20:00 UTC and 14:30 - 21:00 UTC during daylight savings time (summer time).
```
